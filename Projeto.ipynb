```python
#import
import matplotlib.pyplot as plt
from math import *
from cmath import *
import numpy as np
```

```python
def calcula_impedancia(f,sigma):
#     sigma = 20         #parâmetro que varia com a estrutura da antena
    wc = 6.51*(10**6)  #frequência crítica obtida pelo gráfico do item 2
    L = sigma/wc       #indutância indutor
#     print("L:{}".format(L))
    C = 1/(sigma*wc)   #indutancia capacitor
#     print("C:{}".format(C))
    numerador = 1j*f*L*(1/(1j*f*C))
#     print("Numerador:{}".format(numerador))
    denominador = 1j*f*L+(1/(1j*f*C))
#     print("Denominador:{}".format(denominador))
#     print("Divisào:{}".format(numerador/denominador))
    resposta = (numerador/denominador)+ 6*1j*f*L
    return resposta
```

```python
X = np.linspace(1000000, 10000000, 200)

#################################### sigma = 10 ###############################################
Y=[]
for pontos in X:
    y = calcula_impedancia(pontos,10)
    Y.append(np.imag(y))
plt.plot(X,Y)
plt.ylim((-100,100))
plt.plot(8.12*10**6, 0, "or", color="green")
plt.plot(6.51*10**6, 0, "or", color="yellow")
plt.plot(7.0*10**6, 0, "or", color="red")
plt.title("sigma = 10")
plt.xlabel("Frequência")
plt.ylabel("Impedância")
plt.grid()
plt.show()

#################################### sigma = 20 ###############################################
Y=[]
for pontos in X:
    y = calcula_impedancia(pontos,20)
    Y.append(np.imag(y))
plt.plot(X,Y)
plt.ylim((-100,100))
plt.plot(8.12*10**6, 0, "or", color="green")
plt.plot(6.51*10**6, 0, "or", color="yellow")
plt.plot(7.0*10**6, 0, "or", color="red")

plt.grid()
plt.show()

#################################### sigma = 30 ###############################################
Y=[]
for pontos in X:
    y = calcula_impedancia(pontos,30)
    Y.append(np.imag(y))
plt.plot(X,Y)
plt.ylim((-100,100))
plt.plot(8.12*10**6, 0, "or", color="green")
plt.plot(6.51*10**6, 0, "or", color="yellow")
plt.plot(7.0*10**6, 0, "or", color="red")
plt.grid()
np.interp(0, X,Y)
plt.show()

```

```python
def calcula_g(M):
    fr = 8.12*(10**6)
    Rc = 1
    G = (1j*fr*M)/Rc
    return G
```
